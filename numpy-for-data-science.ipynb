# **Numpy array properties:**

```python
#CHECK SHAPE
arr_2d = np.array([[1,2,3],[4,5,6]])
print(arr_2d.shape)
```

```python
#check size
print(arr_2d.size)
```

```python
#NDIM
#number of dimensions
arr_1 = np.array([[1,2,3],[4,5,6]])
arr_2 = np.array([[1,2,3],[4,5,6]])
arr_3 = np.array([[
    [1,2,3],         
    [4,5,6],
    [3,4,5]
]])
print(arr_1.ndim)
print(arr_2.ndim)
print(arr_3.ndim)

```

```python
#dtype data types
arr_1 = np.array([[1,2,3],[4,5,6]])
print(arr_1.dtype)
```
