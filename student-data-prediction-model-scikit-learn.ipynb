```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("devansodariya/student-performance-data")

print("Path to dataset files:", path)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

data = pd.read_csv("/kaggle/input/student-performance-data/student_data.csv")
data.head()
```

```python
X = data[["absences"]]
y = data[["age"]]
print(X,y)
```

```python
model = LinearRegression()
model.fit(X,y)
pr = model.predict(X)
mae = mean_squared_error(y,pr)
mse = mean_squared_error(y,pr)
rmse = np.sqrt(mse)

print(mae)
print(mse)
print(rmse)
```

```python
final_pr = model.predict([[4]])
print(final_pr)
```
